```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_rows = 1000
```

```python
df1 = pd.read_csv('data/page1.csv', header=None)
df2 = pd.read_csv('data/page2.csv', header=None)
df = pd.concat([df1, df2], ignore_index=True)
df.fillna(method='ffill', inplace=True)
for row in df.values:
    row[2:] = [s.replace(',', '.') for s in row[2:]]
df[list(range(2, 11))] = df[list(range(2, 11))].astype(float)
df
```

```python
df.info()
```

```python
values = set()
for row in df.values:
    values.add((row[0], row[1]))
dictionary = dict(enumerate(values))
inv_dict = {v:k for k, v in enumerate(values)}
df['target'] = [inv_dict[(row[0], row[1])] for row in df.values]
df
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.metrics import make_scorer
```

```python
X = df[list(range(2, 11))].values
y = df['target'].values
X, y
```

```python
grid = {'criterion': ['gini', 'entropy'],
        'max_depth': [None],
        'random_state': [42],
        'class_weight': ['balanced', None]}
gs = GridSearchCV(DecisionTreeClassifier(), param_grid=grid, cv=4, scoring=make_scorer(accuracy_score))
gs.fit(X, y)
```
