```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from xml.etree import ElementTree
```

```python
tree = ElementTree.parse('electric power.xml')
```

```python
root = tree.getroot()
```

```python
A = np.empty((len(root), 3), dtype=object)
for i in range(A.shape[0]):
    for j in range(3):
        A[i][j] = root[i][j].text
A
```

```python
df = pd.DataFrame(data=A, columns=['target', 'date', 'hour'])
df
```

```python
df['target'] = df['target'].astype(float)
df['date'] = df['date'].astype('datetime64')
df['hour'] = df['hour'].astype(int)
```

```python
df.plot(x='date', y='target')
```

```python
df_day = df.iloc[:24]
```

```python
df.plot(y='target')
```

```python
data = df.groupby(by='date').sum()
del data['hour']
data
```

```python
plt.figure(figsize=(23, 10))
plt.plot(data['target'])
```

```python
data['day_num'] = range(1,731)
data1 = data[data['day_num'] <= 723]
data1
```

```python
for i in range(1, 8):
    data1['day_' + str(i)] = data1['date'].dt.week == i
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(
                          # Максимальная степень
                          degree=6,
                          # Не генерировать свободный член
                          include_bias=False)
X0 = poly.fit_transform(data1[['day_num']])
X0 = pd.DataFrame(X0, columns=[str(i) for i in range(6)])
X1 = poly.fit_transform(data[['day_num']])
X1 = pd.DataFrame(X1, columns=[str(i) for i in range(6)])
```

```python
model = LinearRegression()
model.fit(X0, data1['target'])
```

```python
plt.figure(figsize=(23, 10))
plt.plot(data1['day_num'], data1['target'])
plt.plot(data['day_num'], model.predict(X1))
plt.show()
```

```python
ans = X1[X1['0'] > 723]
ans2 = data[data['day_num'] > 723]
ans2
```

```python
plt.plot(ans2['day_num'], model.predict(ans))
plt.plot(ans2['day_num'], ans2['target'])
plt.show()
```

```python
print('R^2 для обучающей модели:\t',model.score(X0, data1['target']))
print('R^2 для тестовой модели:\t',model.score(ans, ans2['target']))
```
