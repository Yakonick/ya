```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.options.display.max_rows = 1000

df_red = pd.read_csv("winequality-red.csv", sep=';')
df_white = pd.read_csv("winequality-white.csv", sep=';')
df_white
```

```python
df_red
```

```python
x_red = df_red.drop('quality', axis=1)
y_red = df_red['quality']
x_white = df_white.drop('quality', axis=1)
y_white = df_white['quality']
```

```python
from sklearn.model_selection import train_test_split
x_red_train, x_red_test, y_red_train, y_red_test = train_test_split(x_red, y_red, random_state=42, train_size=0.7)
x_white_train, x_white_test, y_white_train, y_white_test = train_test_split(x_white, y_white, random_state=42, train_size=0.7)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
model = RandomForestClassifier(random_state=42,
                               # опции, относящиеся к отдельным деревьям такаие же, как в tree.DecisionTreeClassifier
                               # число деревьев в лесу
                               n_estimators=20,
                               # функция для impurity ('gini' или 'entropy')
                               criterion='entropy',
                               max_depth=10,
                               # Вычислять out-of-bag ошибку
                               oob_score=True,
                               # использовать результаты предыдущего вызова и нарастить предыдущий лес 
                               warm_start=False,
                               # веса классов для балансировки обучения
                               class_weight=None,#"balanced",
                               max_features="sqrt"
                              )
model.fit(x_red_train, y_red_train)
y_red_pred = model.predict(x_red_test)
print('Accuracy score:\t',metrics.accuracy_score(y_red_pred, y_red_test),'\n')
print(metrics.classification_report(y_red_pred, y_red_test))
```

```python
pd.DataFrame({'feature': x_red.columns,
              'importance': model.feature_importances_}).sort_values('importance', ascending=False)
```

```python
print('Out-of-bag:\t',model.oob_score_)
```

```python
model.fit(x_white_train, y_white_train)
y_white_pred = model.predict(x_white_test)
print('Accuracy score:\t',metrics.accuracy_score(y_white_pred, y_white_test),'\n')
print(metrics.classification_report(y_white_pred,y_white_test))
```

```python
pd.DataFrame({'feature': x_white.columns,
              'importance': model.feature_importances_}).sort_values('importance', ascending=False)
```

```python
print(model.oob_score_)
```

```python
df = pd.concat([df_red, df_white], ignore_index=True)
x = pd.concat([x_red, x_white], ignore_index=True)
y = pd.concat([y_red, y_white], ignore_index=True)

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, train_size=0.7)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print('Accuracy score:\t',metrics.accuracy_score(y_pred, y_test),'\n')
print(metrics.classification_report(y_pred,y_test))
```
